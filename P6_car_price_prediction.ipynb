```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
car_dataset = pd.read_csv('/content/CAR.csv')
```

```python
car_dataset.head()
```

```python
car_dataset.shape
```

```python
car_dataset.info()
```

```python
 car_dataset.isnull().sum()
```

```python
print(car_dataset.fuel.value_counts())
print(car_dataset.seller_type.value_counts())
print(car_dataset.transmission.value_counts())
print(car_dataset.owner.value_counts())
```

Encoding the categorical Data

```python
# encoding "Fuel_type" column
car_dataset.replace({'fuel':{'Petrol':0,'Diesel':1,'CNG':2,'LPG':3,'Electric':4}},inplace=True)

# encoding "Seller_Type" Column
car_dataset.replace({'seller_type':{'Dealer':0,'Individual':1,'Trustmark Dealer':2}},inplace=True)

car_dataset.replace({'transmission':{'Manual':0,'Automatic':1}},inplace=True)
car_dataset.replace({'owner':{'First Owner':0,'Second Owner':1,'Third Owner':2,'Fourth & Above Owner':3,'Test Drive Car':4}},inplace=True)
```

```python
car_dataset.head()
```

Splitting the data and Target

```python
X = car_dataset.drop(['name','selling_price'],axis=1)
Y = car_dataset['selling_price']
```

```python
print(X)
```

```python
print(Y)
```

Splitting Training and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2,random_state=45)
```

Model Training

1.XGBOOST Regression

```python
Regression = XGBRegressor()
```

```python
Regression.fit(X_train,Y_train)
```

Model Evaluation

```python
# prediction on Training data
training_data_prediction = Regression.predict(X_train)
```

```python
# R squared Error
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```

```python
# prediction on Training data
test_data_prediction = Regression.predict(X_test)

error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared Error : ", error_score)
```

Predictive System


```python
input_data = (2025, 0, 0, 4, 0, 0)


# changing the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the data as we are predicting the label for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = Regression.predict(input_data_reshaped)
print(prediction)
```
